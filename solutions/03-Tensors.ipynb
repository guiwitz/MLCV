```python
import imageio.v3 as iio
import torch
import matplotlib.pyplot as plt
```

```python
image = iio.imread('https://github.com/guiwitz/MLCV/raw/main/supp_data/cars_hydrant.jpg')
```

```python
image_tensor = torch.from_numpy(image)
```

```python
image_tensor.dtype
```

```python
image_tensor.requires_grad = True
```

```python
image_tensor = image_tensor.to(torch.float32)
```

```python
image_tensor.requires_grad = True
```

```python
image_flat = torch.flatten(image_tensor)
```
