# Solution: Handling images

````python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np
````

````python
im_ex = iio.imread('https://github.com/guiwitz/microfilm/raw/master/demodata/coli_nucl_ori_ter.tif')
````

````python
im_ex.shape
````

````python
plt.imshow(im_ex[0,0]);
````

````python
import skimage
````

````python
image_filtered = skimage.filters.sobel(im_ex[0,0])
````

````python
plt.imshow(image_filtered);
````

````python
from PIL import Image
````

````python
im_pil = Image.fromarray(im_ex[0,0])
im_pil
````

The ```im_ex``` image has type uint16 so the intensity is scaled for that range of values ($0-2^{16}$). However the maximum value of the image is 181, so we can safely recast it to uint8:

````python
im_ex.max()
````

````python
im_pil = Image.fromarray(im_ex[0,0].astype(np.uint8))
im_pil
````

````python
im_pil.rotate(30)
````

````python
im_pil.rotate(30).save('../output/exercise1.png')
````
